```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_into_num(x):
    token = x.split('-')
    if len(token) == 2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(token[0])
    except:
        return None
```

```python
def removing_price_outliers(df):
    df_out = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        sd = np.std(subdf.price_per_sqft)
        df_filter = subdf[(subdf.price_per_sqft>(m-sd)) & (subdf.price_per_sqft<=(m+sd))]
        df_out = pd.concat([df_out,df_filter],ignore_index = 1)
    return df_out
        
```

```python
df7 = removing_price_outliers(df6)
len(df7),len(df6)
```

```python
def scatter_plot_2_3_bhk(df,location):
    df_2bhk = df[(df.location == location) & (df.bhk_size==2)]
    df_3bhk = df[(df.location == location) & (df.bhk_size==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(df_2bhk.total_sqft,df_2bhk.price,color='blue',marker = '.',label = '2bhk',s = 50)
    plt.scatter(df_3bhk.total_sqft,df_3bhk.price,color='red',marker = '+',label = '3bhk',s = 50)
    plt.xlabel('total_sqft')
    plt.ylabel('price')
    plt.title(location)
    plt.legend()
    
scatter_plot_2_3_bhk(df7,'Rajaji Nagar')
```

```python
def remove_bhk_outliers(df):
    excluded_df = np.array([])
    for loc,loc_df in df.groupby('location'):
        bhk_stats = {}
        for bhk,bhk_df in loc_df.groupby('bhk_size'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count':np.shape(bhk_df.price_per_sqft)[0]
            }
        for bhk,bhk_df in loc_df.groupby('bhk_size'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                excluded_df = np.append(excluded_df,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(excluded_df,axis = 'index')

df8 = remove_bhk_outliers(df7)
df8.head()
            
         
            
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=1) # here random_state is a hyper parameter.
len(X_train),len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=.2 , random_state=0 )
cross_val_score(lr,X,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
steps = [('standard_scaler',StandardScaler()),
        ('linear',LinearRegression())]
pipe1 = Pipeline(steps)

def find_best_search_using_gridsearchCV(X,y):
    algos = {
        'linear_Regression':{
            'model': pipe1['linear'] ,
            'params':{
                'normalize': [True,False]
            }
        },
        'lasso':{
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random','cyclic']
                
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse','friedman_mse'],
                'splitter' : ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits = 5, test_size=.2, random_state=0)
    for algo_name,config in algos.items():
        gs = GridSearchCV(config['model'],config['params'],cv = cv,return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.score(X,y),
            'best_params': gs.best_params_,
        })
    return pd.DataFrame(scores,columns = ['model','best_score','best_params'])

```

```python
find_best_search_using_gridsearchCV(X,y)
```
